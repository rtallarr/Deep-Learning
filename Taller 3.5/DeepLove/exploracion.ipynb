```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
#Load data
tweets = pd.read_csv('data/referenced_tweets_data.csv')
train = pd.read_csv('data/tweets_train.csv')
```

```python
row_num = train.shape[0]
print("Number of rows in train dataset: ", row_num)
```

```python
from random import randint
initial = randint(0, row_num-5)
for i in range(initial, initial+5):
    print("Tweet", i+1)
    print(train["text"].iloc[i])
    print("Odio:", train["Odio"].iloc[i], " "
          "Mujeres:", train["Mujeres"].iloc[i], " "
          "LGBT:", train["Comunidad LGBTQ+"].iloc[i], " ",
          "Migrantes:", train["Comunidades Migrantes"].iloc[i], " ",
          "Pueblos Originarios:", train["Pueblos Originarios"].iloc[i])
    print()
```

```python
# Cantidad de datos por clase
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
gf1 = sns.countplot(x="Mujeres", data=train, ax=axs[0, 0])
gf1.set_ylim(1, 2100)
gf2 =  sns.countplot(x="Comunidad LGBTQ+", data=train, ax=axs[0, 1])
gf2.set_ylim(1, 2100)
gf3 =  sns.countplot(x="Comunidades Migrantes", data=train, ax=axs[1, 0])
gf3.set_ylim(1, 2100)
gf4 =  sns.countplot(x="Pueblos Originarios", data=train, ax=axs[1, 1])
gf4.set_ylim(1, 2100)
```

```python
#create heatmap for all categories 
fig, axs = plt.subplots(figsize=(10, 7))
hm4 = sns.heatmap(train[['Odio', 'Mujeres', 'Comunidad LGBTQ+', 'Comunidades Migrantes', 'Pueblos Originarios']].corr(), annot=True, fmt=".2f")
```
